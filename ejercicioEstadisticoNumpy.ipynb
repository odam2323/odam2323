## **1. Valor esperado**
---

Como parte del equipo de desarrollo del "Caesars Palace", su tarea es emplear sus habilidades de programaci칩n para analizar la integridad de los dados utilizados en el juego  Craps (dados). Este juego se basa en el resultado del lanzamiento de dos dados, sobre el cual los jugadores apuestan; finalmente, los que hayan tenido una mayor proximidad al resultado observado son declarados ganadores en cada ronda, tras la cual podr치n retirar su dinero o apostar nuevamente.

Una de las formas en que la integridad del juego puede verse afectada es la presencia de dados cargados. Con frecuencia, los lanzamientos van desgast치ndolos, haciendo m치s probable que caigan en algunas caras que en otras. Algunos jugadores experimentados son capaces de darse cuenta de este fen칩meno, aprovech치ndolo para apostar por los valores m치s probables y as칤 mejorar su racha de victorias.


Para enfrentar este fen칩meno, el casino ha registrado los resultados de varios lanzamientos de un dado a la vez en m칰ltiples partidas. Esta informaci칩n se le provee en forma de un arreglo de `numpy`, donde cada registro corresponde al resultado en un lanzamiento. **A partir de esto, le solicita que con esta informaci칩n pueda detectar sin un dado est치 cargado o no por medio de la medici칩n del valor esperado.**

En un dado justo y balanceado, el valor esperado se aproxima a 3.5. Si el valor esperado de los dados se desv칤a significativamente de este valor, es probable que est칠 cargado, lo que podr칤a afectar la equidad del juego. Su an치lisis proporcionar치 informaci칩n crucial para asegurar la integridad y la transparencia del casino.

Recuerde que el valor esperado puede calcularce como :

$$ E[x] = \frac{1}{n} \sum x_i $$

donde:
* $E[x]$ corresponde al valor esperado
* $n$ numero de lanzamientos
* $x_i$ el valor obtenido en cada uno de los lanzamientos.

Adicionalmente, el casino le solicita que el c치lculo se realice de la forma m치s eficiente posible, priorizando el rendimiento y la optimizaci칩n del c칩digo. Por esta raz칩n, se solicita evitar el uso de funciones predefinidas de NumPy como `np.mean` y `np.sum`, ya que pueden introducir sobrecarga computacional en este caso particular. En su lugar, se requiere implementar el c치lculo del valor esperado de manera manual, utilizando multiplicaciones matriciales.


**Par치metros**

* `lanzamientos`: arreglo de numpy con los resultados de lanzamientos de 1 a 6.

**Retorna**

* `results`: Valor promedio de los lanzamientos

* La suma de todos los valores de un vector fila se puede obtener realizando el producto matricial por un vector columna de unos.


````
# FUNCI칍N CALIFICADA: valor_esperado(lanzamientos)
def valor_esperado(lanzamientos):
    """
    lanzamientos: un arreglo de NumPy con valores enteros no negativos. Contiene los valores obtenidos en los ultimos 10 lanzamientos.

    Retorna un flotante correspondiente al valor esperado de los lanzamientos.
    """
    ### ESCRIBA SU C칍DIGO AQU칈 ###
    # Reemplace la palabra None por el c칩digo correspondiente
    # ~ 6 l칤neas de c칩digo

    # Convierta los lanzamiento en vector fila

    vector_fila = np.array(lanzamientos)

    # Cree un vector columna de unos con el mismo tama침o de la variable
    # lanzamientos

    vector_columna = np.ones(vector_fila.size)

    # multiplique los vectores obtenidos

    multiplicacion = vector_fila*vector_columna

    resultado = np.sum(multiplicacion)

    # divida el valor obtenido en el tama침o

    esperado = resultado / (vector_fila.size)

    ### FIN DEL C칍DIGO ###

    return esperado
````

## **2. C치lculo de probabilidades**
---
Si bien el proceso anterior permite determinar cu치ndo un dado est치 en condiciones adecuadas para su uso en el juego, esta informaci칩n no puede utilizarse para identificar a los jugadores que aprovecharon esta situaci칩n a su favor. Para ello, el Caesars Palace desea que calcules la probabilidad de ocurrencia de cada cara del dado y utilices esta informaci칩n para compararla con el comportamiento de los jugadores.

Se te proporcionar치 un arreglo de numpy que representa las partidas, donde cada registro corresponde a un lanzamiento. Ahora deber치s determinar la probabilidad de cada cara siguiendo estos pasos:

1. **Contar la frecuencia de cada valor:** Recorre la lista de lanzamientos y cuenta cu치ntas veces aparece cada n칰mero del 1 al 6.
2. **Calcular la probabilidad:** Divide la frecuencia de cada valor entre el n칰mero total de lanzamientos.

游늷 Ejemplo:

Si el n칰mero 3 aparece 20 veces en una lista de 100 lanzamientos, la probabilidad de obtener un 3 ser치:

$$\frac{20}{100} = 0.2 \quad o \quad 20\%$$  

Como resultado del an치lisis, deber치s generar un arreglo llamado probabilidades, donde cada 칤ndice representar치 una de las caras del dado (del 1 al 6). El valor almacenado en cada posici칩n corresponder치 a la probabilidad calculada para esa cara.

* Ejemplo de estructura del arreglo:

    * `probabilidades[2]` contendr치 la probabilidad de obtener un 3.
    * `probabilidades[5]` almacenar치 la probabilidad de obtener un 6

**Par치metros**

* `lanzamientos`: arreglo de numpy con los resultados de lanzamientos de 1 a 6.

**Retorna**

* `probabilidades`: Arreglo con los promedios de los lanzamientos

<details>    
<summary>
    <font size="3" color="darkgreen"><b>Pistas</b></font>
</summary>

* "Para realizar el proceso de conteo de frecuencias, se debe utilizar la indexaci칩n. Cada vez que encuentre un valor en la lista de lanzamientos, utilice dicho valor como 칤ndice para acceder a la posici칩n correspondiente en un arreglo auxiliar y actualice la frecuencia en esa posici칩n.



````
# FUNCI칍N CALIFICADA: calcular_probabilidades(lanzamientos)
def calcular_probabilidades(lanzamientos):
    """
    lanzamientos:  un arreglo de NumPy con valores enteros no negativos. Contiene los valores obtenidos en los ultimos 10 lanzamientos.

    Retorna un arreglo con dimensi칩n (6,) con las probabilidades para cad auna de las caras.
    """
    ### ESCRIBA SU C칍DIGO AQU칈 ###
    # Reemplace la palabra None por el c칩digo correspondiente
    # ~ 1-4 l칤neas de c칩digo

    # Desarrolle un bucle que recorra cada posicion del arreglo lanzamientos.

    vector = np.array(lanzamientos)

    # Change shape to (6,) as required by the docstring

    probabilidades = np.zeros(6)

    # identifique el valor ocurrido y sume 1 a la posicion correspondiente.

    for i in vector:
      probabilidades [i-1] += 1

    # Adjust index to be 0-based

    # Divide by the total number of launches to get probabilities

    probabilidades = probabilidades / np.sum(probabilidades)

    ### FIN DEL C칍DIGO ###

    return probabilidades
````

````
#TEST_CELL
np.random.seed(73)
lanzamientos = np.random.randint(1,7,100)
probabilidades = calcular_probabilidades(lanzamientos)
print(np.sum(probabilidades))
````

## **3. Probabilidades conjuntas**
---
Ahora que es posible determinar la probabilidad de ocurrencia de cada cara, el casino quiere calcular la probabilidad conjunta de que ocurra cada cara del dado y de que un jugador la seleccione en su apuesta. De esta manera, el Caesars Palace podr치 medir la relaci칩n entre las apuestas de un jugador y el comportamiento del dado en la partida.

Para ello, el casino proporciona dos arreglos de numpy de Python:

* Una representa el comportamiento del dado.
* La otra contiene el patr칩n de apuestas de un jugador.

Con estos datos, deber치s calcular:

**Matriz de probabilidades conjuntas:**  Es una matriz en la que cada celda representa la probabilidad observada de que:

* Al arrojar el dado, salga una cara espec칤fica.
* El jugador apueste por esa misma cara.

Caracter칤sticas:

* La suma de las filas representa la probabilidad de ocurrencia de cada cara del dado.
* La suma de las columnas indica la probabilidad de que el jugador apueste por cada cara.

Esta matriz permite analizar la relaci칩n entre el comportamiento del dado y las decisiones del jugador en sus apuestas.

**Matriz de probabilidades ideales:** En esta matriz, cada celda muestra la probabilidad te칩rica de que un n칰mero salga en el dado y que el jugador apueste por 칠l, asumiendo que ambos eventos son independientes.

Se calcula con la siguiente f칩rmula:

$$P(cara_i  \cap apuesta_j) = P(cara_i ) * P(apuesta_j)$$

Donde:

* $P(cara_i   \cap apuesta_j)$ representa la probabilidad conjunta de que al arrojar el dado caiga una cara particular y que el jugador apueste por dicha cara

* $P(cara_i )$ representa la probabilidad de que ocurra la cara i
* $P(apuesta_j)$ representa la probabilidad de que el jugador apueste por la cara j

游늷 Ejemplo:

Si queremos calcular la probabilidad de obtener un '3' en el lanzamiento y que el jugador apueste por la cara '5', debemos multiplicar la probabilidad individual de obtener un '3' al lanzar el dado por la probabilidad individual de obtener un '5' en las apuestas del jugador

$$P(3  \cap 5) = P(3) * P(5)$$

El observar la cercan칤a entre ambas matrices le permitir치 al casino entender que tan asociado est치 el comportamiento del jugador al comportamiento del dado. Entre m치s parecidas sean entre s칤, el jugador tendr치 un comportamiento NO relacionado con el hist칩rico de lanzamiento del dado.



**Par치metros**

* `lanzamientos`: arreglo de numpy con los resultados de lanzamientos de 1 a 6.
* `apuestas`: arreglo de numpy con las apuestas de los jugadores de 1 a 6.

**Retorna**

* `probabilidades_ideales`: Arreglo con las probabilidades te칩ricas
* `probabilidades_medidas`: Arreglo con las probabilidades obtenidas

<details>    
<summary>
    <font size="3" color="darkgreen"><b>Pistas</b></font>
</summary>

* Para obtener la matriz de probabilidades conjuntas de forma eficiente, se puede aprovechar el producto externo. Esta operaci칩n permite multiplicar cada elemento de un vector por todos los elementos de otro vector, generando una matriz con todas las posibles combinaciones de productos.

En este caso, podemos utilizar el producto externo para multiplicar el vector de probabilidades individuales del  dado por el vector de probabilidades de las apuestas. El resultado ser치 una matriz donde cada elemento (i, j) representar치 la probabilidad conjunta de obtener el valor i en el primer dado y el valor j en el segundo dado.



````
# FUNCI칍N CALIFICADA: probabiliades_conjuntas(lanzamientos_1, lanzamientos_2)
def probabiliades_conjuntas(lanzamientos, apuestas):
    """
    lanzamientos:  un arreglo de NumPy con valores enteros no negativos. Contiene los valores obtenidos en los ultimos n lanzamientos del dado.
    apuestas":  un arreglo de NumPy con valores enteros no negativos. Contiene los valores obtenidos en los ultimos n apuestas del jugador.

    Retorna dos matrices con dimensi칩n (6,6) con las probabilidades descritas.
    """
    probabilidades_ideales = None
    probabilidades_medidas = None

    #Calcule la probabiliadad de ocurrencia de las caras para los lanzamientos

    probabilidades_lanzamientos = calcular_probabilidades(lanzamientos)

    #Calcule la probabiliadad de ocurrencia de las caras para las apueats

    probabilidades_apuestas = calcular_probabilidades(apuestas)

    #Calcule la matriz de probabilidades ideales

    probabilidades_ideales = np.outer(probabilidades_lanzamientos,probabilidades_apuestas)

    #En un bucle recorra los avlroes obtenidos para las apuestas y lanzamainetos simulatenamente

    #Cuente e indexe en la matriz de probabilidades conjuntas la cantidad de veces que ocurre la convinaci칩n cara_i y apuesta_j

    probabilidades_medidas = np.zeros((6,6))

    for i,j in zip(lanzamientos, apuestas):
      probabilidades_medidas[i-1][j-1]+=1

    probabilidades_medidas = probabilidades_medidas/np.sum(probabilidades_medidas)


    return probabilidades_ideales, probabilidades_medidas
````

````
## TEST
##############################################
########### SOLUCI칍N (BORRAR) ################
##############################################
def probabiliades_conjuntas(lanzamientos, apuestas):
    """
    lanzamientos:  un arreglo de NumPy con valores enteros no negativos. Contiene los valores obtenidos en los ultimos n lanzamientos del dado.
    apuestas":  un arreglo de NumPy con valores enteros no negativos. Contiene los valores obtenidos en los ultimos n apuestas del jugador.

    Retorna dos matrices con dimensi칩n (6,6) con las probabilidades descritas.
    """
    probabilidades_lanzamientos = calcular_probabilidades(lanzamientos)
    probabilidades_apuestas = calcular_probabilidades(apuestas)

    probabilidades_ideales = probabilidades_lanzamientos @ probabilidades_apuestas

    probabilidades_medidas = np.zeros((6,6))
    for i,j in zip(lanzamientos, apuestas):
        probabilidades_medidas[i-1][j-1]+=1
    probabilidades_medidas = probabilidades_medidas/np.sum(probabilidades_medidas)

    return probabilidades_ideales, probabilidades_medidas
````
